```python
import pandas as pd

persons=pd.read_csv("https://gist.githubusercontent.com/javierIA/7d8564f02ae92bf722800465da2d0f7a/raw/6e0b018296ba1bafe2f20319c9f3e3a8573b151e/salarios.csv")
persons
```

```python
for i in range(5):
  print(persons.sample(frac=1/10),"\n")
```

```python
from sklearn.preprocessing import MinMaxScaler

persons=MinMaxScaler().fit_transform(persons)
persons
```

```python
from sklearn.ensemble import BaggingClassifier
from sklearn.ensemble import BaggingRegressor
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neighbors import KNeighborsRegressor
```

```python
datos=persons.T[:-2].T
clases=persons.T[3]
valor=persons.T[4]
```

```python
clasificador=KNeighborsClassifier(n_neighbors=3)
regresor=KNeighborsRegressor(n_neighbors=3)

numero_de_modelos=90
```

```python
meta_class=BaggingClassifier(base_estimator=clasificador,
                             n_estimators=numero_de_modelos,
                             max_samples=0.30).fit(datos,clases)

meta_regres=BaggingRegressor(base_estimator=regresor,
                             n_estimators=numero_de_modelos,
                             max_samples=0.30).fit(datos,valor)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import r2_score
print("Acc",accuracy_score(clases,meta_class.predict(datos)))
print("Acc",r2_score(valor,meta_regres.predict(datos)))
```
